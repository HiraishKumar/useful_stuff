```python
from yt_dlp import YoutubeDL
import pandas as pd
import matplotlib.pyplot as plt

URL = "https://www.youtube.com/@GlueFactoryPodcast/videos"

ydl_opts = {
    'extract_flat': True,  # Only list video data, no download
    'quiet': True           # Suppress console output
}

with YoutubeDL(ydl_opts) as ydl:
    info = ydl.extract_info(URL, download=False)
    video_titles = [[entry['title'],entry['view_count']] for entry in info['entries'] ]
    print(info.keys())

data_frame = pd.DataFrame(video_titles,columns=["Title","view_count"])

plt.figure(figsize=(12,12))
plt.barh(data_frame['Title'],data_frame['view_count'],color="skyblue")

plt.xlabel("view Count")
plt.ylabel("Video Title")
plt.title('GLUE FACTORY :Video View List')
plt.tight_layout()

plt.show()
```
